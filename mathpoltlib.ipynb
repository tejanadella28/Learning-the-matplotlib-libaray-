# 1.1 Generating Sin and Cos Graph using Matplotlib


```python
x = np.linspace(0 , np.pi ,  10)
y = np.linspace(0,np.pi,20)
```

```python
sin_x = np.sin(x)
cos_y = np.cos(y)
```

```python
plt.plot(x , sin_x , label='sin' , linewidth =2)
plt.plot(y , cos_y , label='cos', linewidth = 2)
plt.legend(loc='lower left')
plt.title('sin and cos waves')
plt.show()
```

# Adjustment plot axes

```python
x = np.linspace(1, 4*np.pi , 10)
sin = np.sin(x)

plt.xlim(0,10) # xlim and ylim are used to change the axes
plt.ylim(-1,1)

plt.plot(x , sin , color ='red', linewidth = 1)
plt.show()
```

# To avoid adjust graph we can use plt.axis('tight') or plt.axis('equal')

```python
x = np.linspace(1, 4*np.pi , 10)
sin = np.sin(x)

plt.axis('equal')

plt.plot(x , sin , color ='red', linewidth = 1)
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)

fig = plt.axes()

fig.plot(x, np.sin(x))

fig.set(xlim = (0,10), ylim = (-2,2), xlabel = 'x', ylabel = 'sin(x)', title = 'Sie Wave')

fig.plot()
```

# Histograms and Binnings

```python
data = np.random.randn(100)
plt.hist(data ,)
plt.show()
```

```python
data = np.random.randn(10000000)

plt.hist(data, bins = 2000, edgecolor = 'blue', alpha = .8, color = 'blue')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

Individual Bins can be removed using histtype='stepfilled'



```python
data = np.random.randn(100)
plt.hist(data, histtype = 'stepfilled' , color = 'red')
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)+2
z = np.random.randn(1000)+4

plt.hist(x , label ='x' , bins=20 , alpha=.7)
plt.hist(y , label='y', bins=20 , alpha =.7)
plt.hist(z , label ='z', bins = 20 , alpha=.7)

plt.show()
```

# Multiple Subplots


```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```
